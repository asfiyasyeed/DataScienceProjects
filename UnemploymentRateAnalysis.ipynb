```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly .express as px
import os
```

```python
data = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.columns = ['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','Longitude','Latitude']
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['States'] = le.fit_transform(data['States'])
data.head()
```

```python
data['Date'] = le.fit_transform(data['Date'])
data.head()
```

```python
data['Frequency'] = le.fit_transform(data['Frequency'])
data.head()
```

```python
data['Region'] = le.fit_transform(data['Region'])
data.head()
```

```python
data.columns = data.columns.str.strip()
print(data.columns)

```

```python
data.corr()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8,6))
ax = sns.heatmap(data.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='w')
plt.show()
```

```python
data.columns=['States', 'Date', 'Estimated Unemployment Rate', 'Estimated Employed',
       'Estimated Labour Participation Rate', 'Region', 'Longitude',
       'Latitude']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

```python
unemployment = data[['States', 'Region', 'Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region','States'],
                     values='Estimated Unemployment Rate',
                     width = 700,height = 600, color_continuous_scale='RdY1Gn',
                     title = "Unemployment Rate in India")
figure.show()
```
